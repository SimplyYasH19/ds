```python
# Assignment 3: Descriptive Statistics 
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
df = iris.frame
```

```python
df['species'] = df['target'].map({
    0: 'Iris-setosa',
    1: 'Iris-versicolor',
    2: 'Iris-virginica'
})

df.drop('target', axis=1, inplace=True)

print("First 5 Rows:\n", df.head())

grouped = df.groupby('species')

print("\n--- Summary Statistics ---")
```

```python
for name, group in grouped:
    print("Mean:\n", group.mean(numeric_only=True))
    print("Median:\n", group.median(numeric_only=True))
    print("Min:\n", group.min(numeric_only=True))
    print("Max:\n", group.max(numeric_only=True))
    print("Standard Deviation:\n", group.std(numeric_only=True))
```

```python
setosa = df[df['species'] == 'Iris-setosa']

print("\n--- Percentiles (Setosa) ---")
print(setosa.describe())

```

```python
# ✅ Theory

# Descriptive statistics is the branch of statistics used to summarize, organize, and describe the 
# important characteristics of a dataset. It provides meaningful insights into the data by using 
# numerical measures and statistical methods. Descriptive statistics helps in understanding the 
# distribution, spread, and central tendency of data.

# The most commonly used measures of central tendency are mean, median, and mode. The mean represents 
# the average value of the dataset, while the median represents the middle value after arranging the 
# data in order. The mode is the value that occurs most frequently in the dataset.

# Measures of variability such as minimum, maximum, range, variance, and standard deviation are used to 
# describe how data values are spread. Standard deviation indicates how far data values deviate from 
# the mean value.

# In this assignment, statistical measures are calculated for different categories of data such as 
# species in the Iris dataset. Grouping the data by categories helps in comparing numerical characteristics 
# among different groups.

# Percentiles and quartiles are also important statistical measures used to understand data distribution. 
# These measures divide the data into equal parts and help identify the position of values within the dataset.

# Descriptive statistics forms the foundation for data analysis, visualization, and machine learning.

# 🔹 Key Points to Write
# Mean, median, and mode
# Standard deviation and variance
# Grouped statistical analysis
# Percentiles and quartiles
# Data distribution analysis

# ✅ Advantages
# Provides quick understanding of data
# Summarizes large datasets effectively
# Helps identify patterns and trends
# Useful for comparison between groups

# ❌ Disadvantages
# Does not predict future outcomes
# Cannot explain relationships between variables
# May oversimplify complex data

# 📌 Applications
# Statistical analysis
# Business reporting
# Research and surveys
# Data analytics and visualization
```
