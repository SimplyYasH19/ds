```python
# Assignment 2: Data Wrangling II 
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [85, 90, np.nan, 120, 95],   # 120 = outlier
    'Attendance': [80, 85, 88, np.nan, 90]
}
```

```python
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)

print("\nAfter Handling Missing Values:\n", df)
```

```python

Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[(df['Marks'] >= lower) & (df['Marks'] <= upper)]
print("\nAfter Removing Outliers:\n", df)
df['Marks_log'] = np.log(df['Marks'])
print("\nAfter Transformation:\n", df)
```
