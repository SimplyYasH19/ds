```python
# Assignment 10: Data Visualization III 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
print("First 5 Rows:\n", df.head())

print("\nFeatures and Data Types:\n", df.dtypes)
```

```python
df.hist(figsize=(8, 6))
plt.suptitle("Feature Distributions (Histogram)")
plt.show()
```

```python
df.plot(kind='box', subplots=True, layout=(2,2), figsize=(8,6))
plt.suptitle("Boxplots for Features")
plt.show()
```
