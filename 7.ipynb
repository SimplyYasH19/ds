```python
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.isalnum() and w not in stop_words]
print("After Stopword Removal:", filtered)
```

```python
pos = nltk.pos_tag(filtered)
print("POS Tags:", pos)
```

```python
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered]
print("Stemmed Words:", stemmed)
```

```python

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]
print("Lemmatized Words:", lemmatized)
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([text])

print("\nTF-IDF Features:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python

# ✅ Theory

# Text Analytics is the process of extracting meaningful information and patterns from textual data 
# using Natural Language Processing (NLP) techniques. Since textual data is unstructured, preprocessing
#  is necessary before analysis.

# The first step is tokenization, where text is divided into smaller units such as words or sentences.
#  Stopword removal is then performed to eliminate commonly used words that do not contribute significant
#  meaning.

# Stemming and lemmatization are used to reduce words to their root or base form. Stemming removes 
# suffixes using simple rules, while lemmatization converts words into meaningful dictionary forms.

# Part-of-Speech (POS) tagging is used to identify grammatical categories such as nouns, verbs, and
#  adjectives. TF-IDF (Term Frequency-Inverse Document Frequency) is then used to represent text 
# numerically by measuring the importance of words in a document relative to the entire collection.

# Text analytics helps process large volumes of text efficiently and supports intelligent NLP applications.

# 🔹 Key Points to Write
# Tokenization
# Stopword removal
# POS tagging
# Stemming and lemmatization
# TF-IDF representation

# ✅ Advantages
# Processes unstructured text data
# Improves NLP model performance
# Efficient text representation
# Extracts meaningful insights

# ❌ Disadvantages
# Complex preprocessing steps
# Loss of contextual meaning
# Computationally expensive for large data

# 📌 Applications
# Sentiment analysis
# Chatbots and virtual assistants
# Search engines
# Text classification systems
```
