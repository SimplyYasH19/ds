```python
# Assignment 8: Data Visualization I 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = sns.load_dataset('titanic')

print("First 5 Rows:\n", df.head())
```

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
plt.hist(df['fare'], bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python

# ✅ Theory

# Data Visualization is the graphical representation of data used to understand patterns, trends, and relationships within datasets. Visualization simplifies complex data and helps in better analysis and decision-making.

# In this assignment, the Titanic dataset is analyzed using visualization libraries such as Matplotlib and Seaborn. A count plot is used to display the number of passengers who survived and did not survive.

# A histogram is also plotted to visualize the distribution of ticket fare values. Histograms divide the data into intervals and show the frequency distribution of numerical variables.

# Visualization helps identify trends, variations, and hidden patterns that may not be easily visible in tabular data. It is an important step in exploratory data analysis.

# 🔹 Key Points to Write
# Graphical representation of data
# Count plot and histogram
# Pattern and trend analysis
# Frequency distribution

# ✅ Advantages
# Easy understanding of data
# Quick identification of trends
# Improves decision making
# Simplifies complex information

# ❌ Disadvantages
# May lead to incorrect interpretation
# Requires proper scaling and labeling
# Large datasets may become cluttered

# 📌 Applications
# Business analytics
# Data exploration
# Dashboard reporting
# Statistical analysis
```
